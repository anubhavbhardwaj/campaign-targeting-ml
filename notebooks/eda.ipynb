### Check Target Distribution

```python
df.target.value_counts(), df.target.value_counts(normalize=True)*100
```

There is mild imbalance in the target. Not as much so to use resampling techniques in the modelling approach for now!

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
labels = ['Group 1 (1)', 'Group 2 (2)', 'Neither (0)']

df.target.value_counts().plot(kind="bar", ax=axes[0], title="Count of Target Classes")
axes[0].set_xticklabels(labels, rotation=0)

df.target.value_counts(normalize=True).plot(kind="bar", ax=axes[1], title="Percentage of Target Classes")
axes[1].set_xticklabels(labels, rotation=0)

plt.show()
```
